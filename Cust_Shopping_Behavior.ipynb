```python
#Loading the dataset using pandas
import pandas as pd
df = pd.read_csv("customer_shopping_behavior.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#Summary statistics
df.describe(include="all")
```

```python
#Checking if missing or Null values are present in the dataset
df.isnull().sum()
```

```python
#Imputing missing values in Review Rating column with the median based on each product category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snakecasing
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns = {'purchase_amount_(usd)' : 'purschase_amount'})
```

```python
df.columns
```

```python
#Create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#Feature Engineering
#creating a new column purchase_frequency_days
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
#checking whether these 2 columns are entirely same
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
#Since both columns are same we will delete 1
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```
